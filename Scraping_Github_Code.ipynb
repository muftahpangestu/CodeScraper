```python
import requests
from bs4 import BeautifulSoup
url = "https://github.com/pallets/flask/blob/master/src/flask/app.py"
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')

```

```python
if "github" in url:
    repo_owner = url.split('/')[3]
repo_owner
```

```python
code_lines = soup.find_all('td', class_="blob-code blob-code-inner js-file-line")
code_raw = []
for item in code_lines:
    line_temp =[]
    for span in item.find_all('span'):
        line_temp.append(span.text)
    code_raw.append(line_temp)

code_raw
    
```

```python
contributors =[]
for item in soup.find_all('a', class_="avatar-link"):
    user_detail = []
    url = item['data-hovercard-url']
    uname = url.split('/')[2]
    user_detail.append(uname)
    user_detail.append("github.com/"+uname)
    contributors.append(user_detail)
contributors
```
