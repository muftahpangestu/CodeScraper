```python
import requests
from bs4 import BeautifulSoup
url = "https://stackoverflow.com/a/472390"
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
url.split('/')[3]
```

```python
if "stackoverflow" in url:
    answer_id = url.split('/')[-1]

print(answer_id)
```

```python
selected_answer = soup.find('div', id = "answer-"+answer_id)
print(selected_answer)
```

```python
answer_info = selected_answer.find_all('div', class_='post-signature grid--cell fl0')[1]


author_details = answer_info.find('div', class_='user-details')
author_uname = author_details.find('a')

time_detail = answer_info.find('div', class_ = 'user-action-time')
time_posted = time_detail.find('span')['title']


reference_detail = [author_uname.text, 'https://stackoverflow.com'+author_uname['href'], time_posted ]
reference_detail
```

```python
selected_code_list = selected_answer.find_all('pre')
combined_code=""
for item in selected_code_list:
    combined_code +=item.text
combined_code
```
