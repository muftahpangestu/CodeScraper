```python
import requests
from bs4 import BeautifulSoup
url = "https://www.programiz.com/python-programming/examples/hello-world"
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
all_pre = soup.find_all('pre')
all_pre
```

```python
Code_list = []
for item in all_pre:
    if item.find('code') != None:
        Code_list.append(item)

Code_list[0].text
```
